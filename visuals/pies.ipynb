```python
import matplotlib.pyplot as plt
```

```python
fake = [68, 20, 6, 6]
neutral = [1, 24, 1, 74]
real = [81, 6, 10, 3]

labels = ['Beide correct', 'Metadata correct & tekst fout', 'Metadata fout & tekst correct', 'Beide fout']
```

```python
fig, axs = plt.subplots(1,3, sharex=True, sharey=True, figsize=(20,6))
fig.patch.set_facecolor('white')
fig.suptitle('Samenvattende plot', size=20)

# plot scores for fake tweets
axs[0].pie(fake, autopct='%.0F%%', textprops={'fontsize':15}, explode=[0.15 for _ in range(4)])
axs[0].set_ylabel('Score (%)', size=15)
axs[0].set_xlabel('Fake', size=15)


# plot scores for neutral tweets
axs[1].pie(neutral, autopct='%.0F%%', textprops={'fontsize':15}, explode=[0.15 for _ in range(4)])
axs[1].set_xlabel('neutral', size=15)

# plot scores for real tweets
axs[2].pie(real, autopct='%.0F%%', textprops={'fontsize':15}, explode=[0.15 for _ in range(4)])
axs[2].set_xlabel('real', size=15)
axs[2].legend(labels, title='Tweet soort:', loc='upper right')

plt.show()

```
